## Exercise 2.7: Collatz sequence

```python
x=103
print(x,end=" ")
while(x!=1):
    if(x%2==0):
        x=x/2
        print(x,end=" ")
    else:
        x=3*x+1
        print(x,end=" ")
```

Write a function that returns the first n primes.

```python
def isPrime(n):
    if n <= 1 :
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def printPrime(n,counter):
        if isPrime(n):
            print(n, end = " ")
            counter+=1
        return counter    

i=2;
n=29
counter=0
while True:
  counter=printPrime(i,counter)
  i+= 1
  if counter==n:
    break
```

## Exercise 3.5: Root finding

Suppose f is a continuous function and f(a) < 0 and f(b) > 0 for some known a and b. For simplicity,
assume a < b. Then, there must exist some c such that f(c) = 0.


Write a function root(f, a, b) that takes a function f and two floats a and b and returns the
root c. Hint: check the sign at the midpoint of the interval.


```python
from scipy.optimize import fsolve
a = lambda x: x**3+10*x
def root(f,a,b):
  fsolve(f, [a, b])

print(root(a,2,80))
```
