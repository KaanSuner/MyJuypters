## Exercise 9.1: Matrix operations

```python
import time
import numpy as np
from scipy.linalg import toeplitz

A = np.random.normal(0.,1.,(200,500))
B = toeplitz(np.array(list(range(1,501))))

adding = A+A
dotproduct1 = A.dot(A.T)
dotproduct2 = A.T.dot(A)
dotproduct3 = A.dot(B)

print("A+A:\n {}\n ".format(adding))
print("A.(A^T):\n {}\n ".format(dotproduct1))
print("(A^T).A:\n {}\n ".format(dotproduct2))
print("A.B:\n {}\n ".format(dotproduct3))

def function(A,B,λ):
  return  A@(B-λ*np.eye(500))

function(A,B,5)
```

##Exercise 9.4: Power iteration


Generate a matrix Z, n × n, with Gaussian entries, and use the power iteration to find the largest eigenvalue and corresponding eigenvector of Z. How many iterations are needed till convergence?


```python
n=200

def powerIteration(Z,exp):
    eigvec = np.ones(n).reshape(n,1)
    max= np.max(eigvec)
    inff = np.inf
    counter = 0

    startTime = time.clock()

    while inff > exp:
        eigvec = Z@eigvec
        max_t = np.max(eigvec)
        inff = np.fabs(max_t-max)
        max = max_t
        eigvec = eigvec / max
        counter += 1
        
    endTime = time.clock()

    return max, counter, endTime-startTime, eigvec

exp = 1e-5
Z = np.random.normal(0,1,(n,n))

print("Maxeig, iterations, time/s, eigvector")
print(powerIteration(Z,exp))
```

##Exercise 9.6: Nearest neighbor

Write a function that takes a value z and an array A and finds the element in A that is closest to z. The function should return the closest value, not index.

```python
def NearestValue(array,z):
  arr = np.asarray(array)
  idx = (np.abs(arr - z)).argmin()
  return arr[idx]
  
array = np.random.random(10)
NearestValue(array,5)
```
